```python
import pandas as pd
import numpy as np
df = pd.read_csv(r"/content/sentimentdataset.csv")
df
```

Mean, Standard Deviation, Variance of LIkes in the Social Media

```python
### Coverting the Dataframe into the list::
total_likes = df['Likes'].to_list()
total_likes
```

```python
### Arraning in the Order ::
total_likes.sort()
total_likes
```

```python
### Maximum and Minimum Values::
max_likes = df["Likes"].max()
print(f"The Maximum Likes is {max_likes}")
min_likes = df["Likes"].min()
print(f"The Minimum Likes is {min_likes}")
```

```python
mean_likes = df["Likes"].sum()/len(total_likes)
print(f"The Mean Likes is {mean_likes}")
sum = 0
for i in total_likes:
  sum += (i-mean_likes)**2
variance_likes = sum/(len(total_likes)-1)
print(f"The Variance Likes of the Social Media Data is {variance_likes}")
standard_deviation = variance_likes** 0.5
print(f"The standard Deviation of the Social Media Data is {standard_deviation}")
```

1st Standard Deviation, 2nd Standard Deviation, 3rd Standard Deviation

```python
first_standard = mean_likes+ 1*(standard_deviation) ### sigma = Standard Deviation
second_standard = mean_likes+ 2*(standard_deviation)
third_standard = mean_likes+ 3*(standard_deviation)
print(f"The First Standard Deviation is {first_standard}")
print(f"The Second Standard Deviation is {second_standard}")
print(f"The Third Standard Deviation is {third_standard}")
```

Calculating the Quartiles Q1,Q2,Q3

```python
### First Quartiles, Q1
### Position of(N+1)/4 items :: 1st Quartiles
### Postion of (N+1)/2 items :: 2nd Quartiles
### Position of 3(N+1)/4 items :: 3rd Quartiles
last_likes = total_likes[-1]
first_likes = total_likes[0]
print(last_likes)
print(first_likes)
pos_q1 = len(total_likes)/4
pos_q2 = len(total_likes)/2
pos_q3 = 3*len(total_likes)/4
print(pos_q1)
print(pos_q2)
print(pos_q3)
```

```python
q1  = total_likes[183-1]
q2 = total_likes[366-1]
q3 = total_likes[549-1]
print(f"The First Quartile is {q1}")
print(f"The Second Quartile is {q2}")
print(f"The Third Quartile is {q3}")
```

Calculating the InterQuartiles Range's

```python
IQR = q3 - q1
print(f"The Inter Quartiles Range is {IQR}")
```

Lower Outliers and Upper Outliers

```python
low_outliers = q1-1.5*IQR
high_outliers = q3+1.5*IQR
print(f"The Lower Outliers are {low_outliers}")
print(f"The Higher Outliers are {high_outliers}")
```
