Sheet 5:: CPI Y-O-Y

### Question ::
Is there a statistically significant difference in Nepal's Consumer Price Index (CPI) inflation between the high inflation period (2022/23) and the current period (2025/26)?

```python
import pandas as pd
import numpy as np
from statsmodels.stats.weightstats import ztest

df1 = pd.read_excel("/content/Current-macroeconomic-situation-Tables-Based-on-Six-Month-data-of-2082.83.xlsx",sheet_name="5.CPI Y-O-Y", header=None, skiprows=7)
df1
```

Cleaning the data of Sheet 5

```python
df1.columns = ['month','idx_2223','pct_2223','idx_2324','pct_2324','idx_2425','pct_2425','idx_2526','pct_2526']
### Delelting the Bad Row's
df1 = df1[df1['month'].notna()]
df1 = df1[df1['month'].astype(str).str.strip() != 'Average']
df1['month'] = df1['month'].astype(str).str.strip()
###Converting all the Columns into number except the
for col in df1.columns[1:]:
  df1[col] = pd.to_numeric(df1[col], errors='coerce')
df1
```

Calculating the Standard Deviation, Mean and the Variance

```python
### Columns 2223 For the Standard Deviation and Mean::
num22_23 = df1['pct_2223'].dropna().to_list()
sample_size1 = len(num22_23)
mean22_23 = df1['pct_2223'].sum()/sample_size1
print(f"The Mean of the 2022/2023 is {mean22_23:2f}")
total22_23 = 0
for i in num22_23:
  total22_23 += (i-mean22_23)**2
variance22_23 = total22_23/(sample_size1)
std22_23 = variance22_23 **0.5
print(f"The Variance of the 2022/2023 is {variance22_23:2f}")
print(f"The Standard Deviation of the 2022/2023 is {std22_23:2f}")
```

```python
### For the Columns 2324 ::
num23_24 = df1['pct_2324'].dropna().to_list()
sample_size2 = len(num23_24)
mean23_24 = df1['pct_2324'].sum()/sample_size2
print(f"The Mean of the 2023/2024 is {mean23_24:2f}")
total23_24 = 0
for i in num23_24:
  total23_24 += (i-mean23_24)**2
variance23_24 = total23_24/(sample_size2)
std23_24 = variance23_24 **0.5
print(f"The Variance of the 2023/2024 is {variance23_24:2f}")
print(f"The Standard Deviation of the 2023/2024 is {std23_24:2f}")
```

```python
### For the Columns:: 2425
num24_25 = df1['pct_2425'].dropna().to_list()
sample_size3 = len(num24_25)
mean24_25 = df1['pct_2425'].sum()/sample_size3
print(f"The Mean of the 2024/2025 is {mean24_25:2f}")
total24_25 = 0
for i in num24_25:
  total24_25 += (i-mean24_25)**2
variance24_25 = total24_25/(sample_size3)
std24_25 = variance24_25 **0.5
print(f"The Vaiance of the 2024/2025 is {variance24_25:2f}")
print(f"The Standard Deviation of the 2024/2025 is {std24_25:2f}")
```

```python
### For the Columns :: 2526
num25_26 = df1['pct_2526'].dropna().to_list()
sample_size4 = len(num25_26)
mean25_26 = df1['pct_2526'].sum()/sample_size4
print(f"The Mean of the 2025/2026 is {mean25_26:2f}")
total25_26 = 0
for i in num25_26:
  total25_26 += (i-mean25_26)**2
variance25_26 = total25_26/(sample_size4)
std25_26 = variance25_26 **0.5
print(f"The Vaiance of the 2025/2026 is {variance25_26:2f}")
print(f"The Standard Deviation of the 2023/2024 is {std25_26:2f}")

```

Mean of the Total Years

```python
means  = [mean22_23,mean23_24,mean24_25,mean25_26]
years = ['2022/23','2023/24','2024/25','2025/26']
sample_size = len(means)
### Calculating the Total mean and Standard Deviation ::
total_mean = 0
for m in means:
  total_mean += m
total_mean = total_mean/sample_size
### for the Standard Deviation::
total_sd = 0
for m in means:
  total_sd += (m-total_mean)**2
total_sd = total_sd/(sample_size)
total_sd = total_sd**0.5
print(f"The Mean of the Overall Group is {total_mean:2f}")
print(f"The Standard Deviation of the Overal Group is {total_sd:2f}")
```

Z Score of the Total Year

```python
## Z score::
for i in range(len(years)):
  z = (means[i]-total_mean)/total_sd
  print(f"The Z score of {years[i]} is {z:2f}")
```

Z Test of 2022/2023 Vs 2024/2025

```python
pct_2223 = df1['pct_2223'].dropna().to_list()
pct_2526 = df1['pct_2526'].dropna().to_list()
z_stat, p_value = ztest(pct_2223,pct_2526)
print(f"The Z Score is {z_stat:2f}")
print(f"The P Value is {p_value:2f}")
if p_value < 0.05:
  print("Reject the Null Hypothesis")
else:
  print("Accept the Null Hypothesis")
```

Z Test of 2022/2023 Vs 2024/25

```python
pct_2223 = df1['pct_2223'].dropna().to_list()
pct_2425 = df1['pct_2425'].dropna().to_list()
z_stat, p_value = ztest(pct_2223,pct_2425)
print(f"The Z Score is {z_stat:2f}")
print(f"The P Value is {p_value:2f}")
if p_value < 0.05:
  print("Reject the Null Hypothesis")
else:
  print("Accept the Null Hypothesis")
```

## Conclusion::
Nepal’s CPI inflation has significantly decreased from 2022/23 to 2025/26. Both Z-tests confirm that the reductions between 2022/23 vs 2024/25 and 2022/23 vs 2025/26 are statistically significant (p < 0.05), indicating a strong and consistent downward trend in inflation.

```python
import matplotlib.pyplot as plt

months = ['Aug','Sep','Oct','Nov','Dec','Jan','Feb','Mar','Apr','May','Jun','Jul']

plt.figure(figsize=(12,5))
plt.plot(months, df1['pct_2223'], marker='o', label='2022/23', color='red')
plt.plot(months, df1['pct_2324'], marker='o', label='2023/24', color='orange')
plt.plot(months, df1['pct_2425'], marker='o', label='2024/25', color='green')
plt.plot(months[:6], df1['pct_2526'].dropna(), marker='o', label='2025/26 (6m)', color='blue')

plt.title("Nepal CPI Inflation — Monthly % Change")
plt.xlabel("Month")
plt.ylabel("CPI % Change")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## What the Chart Say Conclusion::
The monthly CPI trend shows fluctuations in inflation across different months.
Inflation decreases in the later months compared to earlier periods.
Despite short-term variations, the overall trend is downward.
This indicates that inflation is gradually declining over time
