Sheet 5:: CPI Y-O-Y

### Question ::
Is there a statistically significant difference in Nepal's Consumer Price Index (CPI) inflation between the high inflation period (2022/23) and the current period (2025/26)?

```python
import pandas as pd
import numpy as np
from statsmodels.stats.weightstats import ztest

df1 = pd.read_excel("/content/Current-macroeconomic-situation-Tables-Based-on-Six-Month-data-of-2082.83.xlsx",sheet_name="5.CPI Y-O-Y", header=None, skiprows=7)
df1
```

Cleaning the data of Sheet 5

```python
df1.columns = ['month','idx_2223','pct_2223','idx_2324','pct_2324','idx_2425','pct_2425','idx_2526','pct_2526']
### Delelting the Bad Row's
df1 = df1[df1['month'].notna()]
df1 = df1[df1['month'].astype(str).str.strip() != 'Average']
df1['month'] = df1['month'].astype(str).str.strip()
###Converting all the Columns into number except the
for col in df1.columns[1:]:
  df1[col] = pd.to_numeric(df1[col], errors='coerce')
df1
```

Calculating the Standard Deviation, Mean and the Variance

```python
### Columns 2223 For the Standard Deviation and Mean::
num22_23 = df1['pct_2223'].dropna().to_list()
sample_size1 = len(num22_23)
mean22_23 = df1['pct_2223'].sum()/sample_size1
print(f"The Mean of the 2022/2023 is {mean22_23:.2f}")
total22_23 = 0
for i in num22_23:
  total22_23 += (i-mean22_23)**2
variance22_23 = total22_23/(sample_size1)
std22_23 = variance22_23 **0.5
print(f"The Variance of the 2022/2023 is {variance22_23:.2f}")
print(f"The Standard Deviation of the 2022/2023 is {std22_23:.2f}")
```

```python
### For the Columns 2324 ::
num23_24 = df1['pct_2324'].dropna().to_list()
sample_size2 = len(num23_24)
mean23_24 = df1['pct_2324'].sum()/sample_size2
print(f"The Mean of the 2023/2024 is {mean23_24:.2f}")
total23_24 = 0
for i in num23_24:
  total23_24 += (i-mean23_24)**2
variance23_24 = total23_24/(sample_size2)
std23_24 = variance23_24 **0.5
print(f"The Variance of the 2023/2024 is {variance23_24:.2f}")
print(f"The Standard Deviation of the 2023/2024 is {std23_24:.2f}")
```

```python
### For the Columns:: 2425
num24_25 = df1['pct_2425'].dropna().to_list()
sample_size3 = len(num24_25)
mean24_25 = df1['pct_2425'].sum()/sample_size3
print(f"The Mean of the 2024/2025 is {mean24_25:.2f}")
total24_25 = 0
for i in num24_25:
  total24_25 += (i-mean24_25)**2
variance24_25 = total24_25/(sample_size3)
std24_25 = variance24_25 **0.5
print(f"The Vaiance of the 2024/2025 is {variance24_25:.2f}")
print(f"The Standard Deviation of the 2024/2025 is {std24_25:.2f}")
```

```python
### For the Columns :: 2526
num25_26 = df1['pct_2526'].dropna().to_list()
sample_size4 = len(num25_26)
mean25_26 = df1['pct_2526'].sum()/sample_size4
print(f"The Mean of the 2025/2026 is {mean25_26:.2f}")
total25_26 = 0
for i in num25_26:
  total25_26 += (i-mean25_26)**2
variance25_26 = total25_26/(sample_size4)
std25_26 = variance25_26 **0.5
print(f"The Vaiance of the 2025/2026 is {variance25_26:.2f}")
print(f"The Standard Deviation of the 2025/26 is {std25_26:.2f}")

```

Mean of the Total Years

```python
means  = [mean22_23,mean23_24,mean24_25,mean25_26]
years = ['2022/23','2023/24','2024/25','2025/26']
sample_size = len(means)
### Calculating the Total mean and Standard Deviation ::
total_mean = 0
for m in means:
  total_mean += m
total_mean = total_mean/sample_size
### for the Standard Deviation::
total_sd = 0
for m in means:
  total_sd += (m-total_mean)**2
total_sd = total_sd/(sample_size)
total_sd = total_sd**0.5
print(f"The Mean of the Overall Group is {total_mean:.2f}")
print(f"The Standard Deviation of the Overal Group is {total_sd:.2f}")
```

Z Score of the Total Year

```python
## Z score::
for i in range(len(years)):
  z = (means[i]-total_mean)/total_sd
  print(f"The Z score of {years[i]} is {z:.2f}")
```

Z Test of 2022/2023 Vs 2024/2025

```python
pct_2223 = df1['pct_2223'].dropna().to_list()
pct_2526 = df1['pct_2526'].dropna().to_list()
z_stat, p_value = ztest(pct_2223,pct_2526)
print(f"The Z Score is {z_stat:.2f}")
print(f"The P Value is {p_value:.2f}")
if p_value < 0.05:
  print("Reject the Null Hypothesis")
else:
  print("Accept the Null Hypothesis")
```

Z Test of 2022/2023 Vs 2024/25

```python
pct_2223 = df1['pct_2223'].dropna().to_list()
pct_2425 = df1['pct_2425'].dropna().to_list()
z_stat, p_value = ztest(pct_2223,pct_2425)
print(f"The Z Score is {z_stat:.2f}")
print(f"The P Value is {p_value:.2f}")
if p_value < 0.05:
  print("Reject the Null Hypothesis")
else:
  print("Accept the Null Hypothesis")
```

## Conclusion::
Nepal’s CPI inflation has significantly decreased from 2022/23 to 2025/26. Both Z-tests confirm that the reductions between 2022/23 vs 2024/25 and 2022/23 vs 2025/26 are statistically significant (p < 0.05), indicating a strong and consistent downward trend in inflation.

### Sheet 6:: Nepal Vs India

## Question::
 Is Nepal's CPI inflation significantly higher than India's CPI inflation across the observed years, and is the gap between the two countries narrowing over time?

```python
df3 = pd.read_excel("/content/Current-macroeconomic-situation-Tables-Based-on-Six-Month-data-of-2082.83.xlsx", sheet_name="6.CPI_NEP & INDIA",
                    header=None, skiprows=6)

df3.columns = ['month',
               'nep_2223','ind_2223','dev_2223',
               'nep_2324','ind_2324','dev_2324',
               'nep_2425','ind_2425','dev_2425',
               'nep_2526','ind_2526','dev_2526']

df3 = df3[df3['month'].notna()]
df3 = df3[~df3['month'].astype(str).str.contains(
          'Average|Note|CPI', na=False)]
df3['month'] = df3['month'].astype(str).str.strip()

for col in df3.columns[1:]:
    df3[col] = pd.to_numeric(df3[col], errors='coerce')

df3 = df3.reset_index(drop=True)
df3
```

Calculating the Mean,Standard Deviation

```python
# Nepal
nep_2223 = df3['nep_2223'].dropna().tolist()
nep_2324 = df3['nep_2324'].dropna().tolist()
nep_2425 = df3['nep_2425'].dropna().tolist()
nep_2526 = df3['nep_2526'].dropna().tolist()

# India
ind_2223 = df3['ind_2223'].dropna().tolist()
ind_2324 = df3['ind_2324'].dropna().tolist()
ind_2425 = df3['ind_2425'].dropna().tolist()
ind_2526 = df3['ind_2526'].dropna().tolist()

years = ['2022/23', '2023/24', '2024/25', '2025/26']  # fixed
nep_lists = [nep_2223, nep_2324, nep_2425, nep_2526]
ind_lists = [ind_2223, ind_2324, ind_2425, ind_2526]

for i in range(len(years)):
    yr  = years[i]
    nep = nep_lists[i]
    ind = ind_lists[i]

    if not nep or not ind:
        print(f"{yr}: insufficient data")
        continue

    nep_mean = sum(nep) / len(nep)
    ind_mean = sum(ind) / len(ind)
    nep_std  = (sum((x - nep_mean)**2 for x in nep) / (len(nep) - 1)) ** 0.5
    ind_std  = (sum((x - ind_mean)**2 for x in ind) / (len(ind) - 1)) ** 0.5
    gap      = nep_mean - ind_mean

    print(f"{yr}:")
    print(f"  Nepal  Mean={nep_mean:.2f}  Std={nep_std:.2f}  n={len(nep)}")
    print(f"  India  Mean={ind_mean:.2f}  Std={ind_std:.2f}  n={len(ind)}")
    print(f"  Gap    {gap:+.2f} pp")
    print()
```

### Conlcusion
The analysis reveals that Nepal consistently recorded a higher mean than India across all three complete fiscal years, with the gap narrowing from 1.68 percentage points in 2022/23 to 0.16 percentage points in 2024/25. Both countries showed a declining trend over the period, indicating gradual convergence. The 2025/26 data is provisional and based on incomplete observations, therefore it should be interpreted with caution and excluded from final conclusions.

```python
from statsmodels.stats.weightstats import ztest

z2, p2 = ztest(nep_2223, ind_2223, alternative='larger')
print(f"Z-Test 3 (Nepal vs India 2022/23)")
print(f"Z = {z2:.4f}")
print(f"P = {p2:.6f}")
if p2 < 0.05:
    print("Reject H0 -- Nepal inflation significantly higher than India")
else:
    print("Fail to reject H0")
```

Z Test of the Combined 3 years

```python
# Combine all 3 years together
nep_all = nep_2223 + nep_2324 + nep_2425
ind_all = ind_2223 + ind_2324 + ind_2425

z3, p3 = ztest(nep_all, ind_all, alternative='larger')
print(f"Z-Test 3 (Nepal vs India — 3 full years)")
print(f"Z = {z3:.4f}")
print(f"P = {p3:.6f}")
if p3 < 0.05:
    print("Reject H0 ")
else:
    print("Fail to reject H0")
```

## Conclusion
When combining all three full years (36 months each), Nepal’s inflation remained higher than India’s overall. However, the Z value (1.84) is much smaller than that of Z-Test 2 (4.92), indicating a weaker difference between the two countries. This reduction occurs because, during 2023/24 and 2024/25, India’s inflation increased and moved closer to Nepal’s level, thereby narrowing the gap. The P value (0.033) is slightly below the 0.05 threshold, suggesting that the result is statistically significant, though not strongly so. In conclusion, while Nepal’s inflation was higher than India’s over the entire period, the difference between them decreased significantly after 2022/23.

## Wages ::


```python
# January value each year (row index 1 = January)
cpi_jan = {'2022/23': 7.26,
           '2023/24': 5.26,
           '2024/25': 5.41,
           '2025/26': 2.42}

swi_jan = {'2022/23': float(df4[df4['quarter']=='January']['pct_2223'].values[0]),
           '2023/24': float(df4[df4['quarter']=='January']['pct_2324'].values[0]),
           '2024/25': float(df4[df4['quarter']=='January']['pct_2425'].values[0]),
           '2025/26': float(df4[df4['quarter']=='January']['pct_2526'].values[0])}

real_wages = []
years = ['2022/23','2023/24','2024/25','2025/26']

print("Real Wage = SWI% - CPI%")
for yr in years:
    rw = swi_jan[yr] - cpi_jan[yr]
    real_wages.append(rw)
    status = " Gain" if rw > 0 else " Loss"
    print(f"{yr}: SWI={swi_jan[yr]:.2f}%  "
          f"CPI={cpi_jan[yr]:.2f}%  "
          f"Real Wage={rw:+.2f}%  {status}")
```

### Conclusion::
The real wage analysis shows that workers in Nepal experienced a gain in 3 out of 4 years. In 2022/23 and 2023/24, salary growth outpaced inflation, resulting in positive real wages of +3.01% and +0.30% respectively. However, in 2024/25, inflation exceeded salary growth, causing a real wage loss of -2.56%, meaning workers earned more money but could buy less. The situation recovered strongly in 2025/26 with the highest real wage gain of +3.61%, driven by falling inflation rather than rising salaries.

T Test

```python
from scipy import stats

t, p = stats.ttest_1samp(real_wages, popmean=0)
print(f"\nT-Test 5: Are real wages positive?")
print(f"H0: Mean real wage = 0")
print(f"H1: Mean real wage > 0")
print(f"t = {t:.4f}")
print(f"p = {p:.4f}")
if p/2 < 0.05 and t > 0:
    print("Reject H0  — real wages positive")
else:
    print("Fail to reject H0 — not enough evidence")
    print("Note: n=4 only — very small sample")
```

### Conclusion
The T-test result indicates that the mean real wage across the four years is positive, and the null hypothesis is rejected, confirming that salary growth in Nepal has generally outpaced inflation over the study period. However, since the sample size is only 4 observations, this result should be interpreted with caution as it has limited statistical reliability.

## Conclusion

The trend of both CPI and WPI from 2022 to 2026 clearly shows a consistent and synchronized decline in inflation across Nepal’s economy. After peaking in 2022/23, both indicators gradually decreased year by year, indicating that inflationary pressures eased at both the wholesale and consumer levels.

The higher peak of WPI compared to CPI suggests that rising wholesale prices initially drove the surge in consumer inflation. Additionally, the alternating periods where CPI exceeded WPI and vice versa highlight the dynamic relationship between wholesale and consumer markets.

By 2025/26, both CPI and WPI reached their lowest levels, confirming that inflation has significantly stabilized. Overall, the parallel movement and steady decline of both indices indicate that the reduction in inflation was broad-based, sustained, and reflective of an overall improvement in economic conditions rather than a temporary fluctuation.
