Gym Members Exercise Dataset

```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/gym_members_exercise_tracking.csv')
df.head()
```

```python
df.describe()
```

Convering the  weight and height dataset in the List

```python
## Weight in the Kg::
weight = df['Weight (kg)'].tolist()
## Height in the Meter::
height = df['Height (m)'].tolist()
```

Couting the Number of the Sample Size

```python
sample_size1 = len(df)
print(f"The Sample Size of the Height and height is {sample_size1}")
```

Calculating the Mean, Variance, and the Standard deviation of Weight amd Height

```python
### Mean of the Weight::
mean_weight = sum(df['Weight (kg)']) / len(df['Weight (kg)'])
print(mean_weight)
### Mean of the Height ::
mean_height = sum(df['Height (m)']) / len(df['Height (m)'])
print(mean_height)
```

```python
### Standard Deviation of the weight
sum_weight = 0
for i in weight:
  sum_weight += (i - mean_weight) ** 2
variance_weight = sum_weight / len(weight)
print(f"The Variance of the Weight is {variance_weight}")
std_weight = variance_weight ** 0.5
print(f"The Standard Deviation of the Weight is {std_weight}")
### Standard Deviation of the Height
sum_height = 0
for i in height:
  sum_height += (i - mean_height) ** 2
variance_height = sum_height / len(height)
print(f"The Variance of the height is {variance_height}")
std_height = variance_height ** 0.5
print(f"The Standard Deviation of the height is {std_height}")
```

Calculating the Standard Error of Height and Weight

```python
### For the Height
standard_error_height = std_height / (sample_size1 ** 0.5)
print(f"The Standard Error of the Height is {standard_error_height}")
### For the Weight ::
standard_error_weight = std_weight / (sample_size1 ** 0.5)
print(f"The Standard Error of the Weight is {standard_error_weight}")
```

**Conclusion **

The standard error of height (0.0040) is very small, which means the mean height is estimated very accurately. The standard error of weight (0.6795) is slightly larger, showing more variability in weight, but the mean weight estimate is still reasonably reliable due to the large sample size.

### Pearson's Correlation using Formula in Python

```python

### Calculating the Sum needed for the Correlation
sum_cov = 0
sum_height_sq = 0
sum_weight_sq = 0

for i in range(sample_size1):
    diff_height = height[i] - mean_height
    diff_weight = weight[i] - mean_weight

    sum_cov += diff_height * diff_weight
    sum_height_sq += diff_height**2
    sum_weight_sq += diff_weight**2

### Calculating the Correlation Pearosn's using formula
pearson_r = sum_cov / ( (sum_height_sq**0.5) * (sum_weight_sq**0.5) )

print("Pearson Correlation Height & Weight:", pearson_r)
```

### Conclusion::

The Pearson correlation between height and weight is approximately 0.37, which indicates a moderate positive relationship. This means that, in general, as height increases, weight tends to increase as well, but the relationship is not very strong.
