Student Exam Scores dataset

```python
import pandas as pd
import numpy as np
df = pd.read_csv("/content/student_exam_scores.csv")
df.head()
```

```python
df.describe()
```

Converting the exam_score  item's into list and counting

```python
exam_score = df['exam_score'].tolist()
sample_size = len(exam_score)
print(f"The Sample Size of the Dataset is {sample_size}")
```

Mean of the Score exam

```python
mean = df['exam_score'].sum()/sample_size
print(f"The Mean of the Exam Score is {mean}")
```

Calculating the Standard Deviation and Variance

```python
sum = 0
for i in exam_score:
  sum = sum + ((i-mean)**2)
variance = sum/(sample_size-1)
print(f"The Variance of the Exam Score is {variance}")
standard_deviation = variance**0.5
print(f"The Standard Deviation of the Exam Score is {standard_deviation}")
```

Calculating the Standard Error of the Exam Scores

```python
std_error = standard_deviation/(sample_size**0.5)
print(f"The Standard Error of the Exam Score is {std_error}")
```

## Conclusion of Standard Error::
The standard error is 0.480, which shows that the calculated mean exam score is a reliable estimate of the population mean

### Calculating the Margin of Error of Exam Score
For this analysis, we have chosen a 95% confidence level.
The corresponding Z-value from the standard normal distribution (Z-table) is 1.96.

```python
z = 1.96
ME = z * std_error
print(f"The Margin of Error of Exam Score is {ME}")
```

### Conclusion of Margin of Error
The Margin of Error (ME) of the exam scores is 0.941. This means that the sample mean exam score could vary by ±0.941 from the true population mean with 95% confidence.

### Calculating the Confident Interval(C.I)  of Exam Score

```python
### Lower Limit of Confident Interval
CI_lower = mean - ME
## Upper limit of confident Interval
CI_upper = mean + ME
print(f"95% Confidence Interval: [{CI_lower}, {CI_upper}]")
```

## Conclusion::
The 95% confidence interval of exam scores is 33.01 to 34.90, indicating the sample mean is a reliable estimate of the true population mean.
