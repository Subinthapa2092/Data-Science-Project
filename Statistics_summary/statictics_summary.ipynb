```python
data = [19,29,40,22,9,12,18,100,0]
# # sum_value = sum(data)
# print(data)
# print(sum_value)
df = pd.DataFrame(data)
df.describe()
data.sort()
print(data)
df.describe()
```

Mean, Standard Deviation , and Vairance

```python
### Mean Values
mean_value = sum(data)/len(data)
print(mean_value)
```

```python
### Standard Deviation and Variance
total = 0
for i in data:
  total += (i - mean_value)**2
variance = total/(len(data)-1) ### Sample Variance
standard_deviation = variance ** 0.5
print(f"The Standared Deviation of the Given Sample is  {standard_deviation}")
print(f"The Variance of the Given Sample is {variance}")
```

Calculating the First, Second and the third Standard Deviation Taking Positive and Negative

```python
first_std = mean_value + standard_deviation
first_std1  = mean_value - standard_deviation
second_std = mean_value + (2*standard_deviation)
second_std1 = mean_value - (2*standard_deviation)
third_std = mean_value + (3*standard_deviation)
third_std1 = mean_value - (3*standard_deviation)
print(f"The First Standard Deviation is {first_std1} to {first_std}")
print(f"The Second Standard Deviation is {second_std1} to {second_std}")
print(f"The Third Standard Deviation is {third_std1} to {third_std}")
```

Calculating the Quartiles of the Data Sample with q1,q2, q3

```python
### Position of the Q1, Q2, Q3
post_q1 = (len(data)+1)/4
post_q2 = (len(data)+1)/2
post_q3 = 3*(len(data)+1)/4
print(post_q1)
print(post_q2)
print(post_q3)
```

Interquartile Range

```python
iqr = q3 - q1
print(f"The Inter Quartile Range is {iqr}")
```

Finding the Outliers range and Couting it's

```python
## Upper OUtliers
out_upper = q3+1.5*iqr
out_upper
### Lower Outliers
out_lower = q1-1.5*iqr
out_lower
print(f"The Outliers is From the {out_lower} from {out_upper}")
count = 0
for i in data:
  if i > out_upper or i < out_lower:
    count += 1
print(f"The Number of Outliers is {count}")
```

```python
### 2nd Central Moment
second_sum  = 0
for i in data:
  second_sum += (i-mean_value)**2
cental_moment2 = second_sum/(len(data)-1) ## Sample Variance
print(cental_moment2)
```

```python
### 3rd Cental Moment
third_sum  = 0
for i in data:
  third_sum += (i-mean_value)**3
cental_moment3 = third_sum/(len(data)-1) ## Sample Variance
print(cental_moment3)
```

```python
### 4th Cental Moment
fourth_sum = 0
for i in data:
  fourth_sum += (i-mean_value)**4
cental_moment4 = fourth_sum/(len(data)-1)
print(cental_moment4)
```
