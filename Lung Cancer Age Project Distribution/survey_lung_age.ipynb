Survey Lung Cancer::

```python
import pandas as pd
import numpy as np
df = pd.read_csv("/content/survey lung cancer.csv")
df.head(5)
```

```python
df.describe()
```

```python
age = df["AGE"].to_list()
age.sort()
print(age)
```

Calculating the Mean, Variance, Standard Deviation

```python
### Mean ::
mean = df["AGE"].sum()/len(df["AGE"])
print(mean)
```

```python
### Variance
sum = 0
for i in df["AGE"]:
  sum += (i-mean)**2
variance = sum /len(df["AGE"])
print(variance)
### Standard Deviation ::
std = variance ** 0.5
print(std)
```

Positon of the Quartiles and it's Value's

```python
pos_q1 = (len(age)+1)/4 ### Lower Index
pos_q2  = (len(age)+1)/2
pos_q3 = 3*(len(age)+1)/4 ### Upper Index
print(pos_q1)
print(pos_q2)
print(pos_q3)
```

```python
# Q1 interpolation
q1 = age[int(pos_q1)-1] + (pos_q1 - int(pos_q1)) * (age[int(pos_q1)] - age[int(pos_q1)-1])

# Q2 (median)
q2 = age[int(pos_q2)-1]  # no interpolation if integer

# Q3 interpolation
q3 = age[int(pos_q3)-1] + (pos_q3 - int(pos_q3)) * (age[int(pos_q3)] - age[int(pos_q3)-1])
print(q1)
print(q2)
print(q3)
```

Inter Quartiles Range

```python
iqr = q3-q1
print(iqr)
```

Calculating the Standard Deviation with different types::

```python
### Taking the Positive Part
sd_pos1 = mean + std
sd_pos2 = mean + 2*std
sd_pos3 = mean + 3*std
### Taking the Negative Part
sd_neg1 = mean - std
sd_neg2 = mean - 2*std
sd_neg3 = mean -3*std
print(f"The First Standard Deviation is {sd_neg1} to {sd_pos1}")
print(f"The Second Standard Deviation is {sd_neg2} to {sd_pos2}")
print(f"The Third Standard Deviation is {sd_neg3} to {sd_pos3}")
```

Finding the Outiers range and the couting it's

```python
low_outlier = q1-1.5*iqr
high_outlier = q3+1.5*iqr
print(f"The Outlier Range is below {low_outlier} to above {high_outlier}")
```

```python
### Couting the Outlier's
count = 0
for i in df["AGE"]:
  if i <= low_outlier or i >= high_outlier:
    count += 1
print(f"The total Number of the Outliers is {count}")
```

Central Moment's and it's Types

```python
### 1st Central Moment
first_moment = 0
for i in df["AGE"]:
  first_moment += (i-mean)**1
first_moment = first_moment/len(df["AGE"])
print(f"The First Moment is {first_moment}") ## It is assumed to be Zero(0)
```

```python
## 2nd Central MOment
second_moment = 0
for i in df['AGE']:
  second_moment += (i-mean)**2
second_moment = second_moment/len(df['AGE'])
print(f"The Second Moment is {second_moment}")
```

```python
### 3rd Central Moment
third_moment = 0
for i in df['AGE']:
  third_moment += (i-mean)**2
third_moment = third_moment/len(df['AGE'])
print(f"The third Moment is {third_moment}")
```

```python
### 4th Central Moment
fourth_moment = 0
for i in df["AGE"]:
  fourth_moment += (i-mean)**4
fourth_moment = fourth_moment/len(df["AGE"])
print(f"The fourth Moment is {fourth_moment}")
```

```python

if third_moment == 0:
  print("Symmetric")
elif third_moment > 0:
  print("Right Skewed")
else:
  print("Left Skewed")
```

Conclusion::
### Right Skewed (Positive Skew) → Most values are below the mean; tail stretches to the right.

Making the Visual of the Distribution

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
### Making the distributio
x = np.linspace(mean - 3*std, mean + 3*std, 100)
plt.plot(x,norm.pdf(x,mean,std))
### Making the distributio
x = np.linspace(mean - 3*std, mean + 3*std, 100)
plt.title('Normal Distribution')
plt.xlabel('AGE')
plt.ylabel('Probability')
plt.show()
```

Conclusion: The age distribution of patients is roughly bell-shaped and almost symmetric, centered around 60–65 years. Most patients fall within this age range, with a few older patients creating a slight right skew. Extreme young or very old ages are rare.
